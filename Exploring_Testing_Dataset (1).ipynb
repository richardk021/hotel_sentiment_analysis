```python
texts = data['Review_lits'].tolist()
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

lda_model = models.ldamodel.LdaModel(corpus=corpus,id2word=dictionary,num_topics=10,passes=10)

for i, topic in lda_model.show_topics(num_topics=10, num_words=10, formatted=False):
    print(f"Topic {i+1}: {' '.join([word[0] for word in topic])}")
```

```python
for i, topic in lda_model.show_topics(num_topics=10, num_words=10, formatted=False):
    print(f"Top influential words for Topic {i+1}:")
    influential_words = sorted(topic, key=lambda x: x[1], reverse=True)[:10]
    for word, weight in influential_words:
        print(f"{word}: {weight:.4f}")
    print("\n")
```

```python

texts = data['Review_lits'].tolist()
sentiments = data['KNN Predict'].tolist()

positive_texts = [texts[i] for i in range(len(texts)) if sentiments[i] == 'Positive']
neutral_texts = [texts[i] for i in range(len(texts)) if sentiments[i] == 'Neutral']
negative_texts = [texts[i] for i in range(len(texts)) if sentiments[i] == 'Negative']

def train_lda_and_plot(texts, sentiment_label):
    if not texts:
        print(f"No texts found for {sentiment_label} sentiment.")
        return

    # Membuat dictionary dan corpus
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]

    if not corpus:
        print(f"No terms found for {sentiment_label} sentiment.")
        return

    # Melatih model LDA
    lda_model = models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=10, passes=10)

    # Menampilkan topik-topik dan kata-kata dengan bobot tertinggi dalam plot
    top_n_words = 10
    for i, topic in lda_model.show_topics(num_topics=10, num_words=top_n_words, formatted=False):
        words, weights = zip(*topic)
        plt.figure(figsize=(10, 5))
        plt.barh(range(len(words)), weights, align='center')
        plt.yticks(range(len(words)), words)
        plt.gca().invert_yaxis()
        plt.xlabel('Weight')
        plt.title(f'Topic {i+1} for {sentiment_label} sentiment')
        plt.show()

# Melatih model dan membuat plot untuk setiap sentimen
print("\nPositive\n")
train_lda_and_plot(positive_texts, 'Positive')
print("\nNeutral\n")
train_lda_and_plot(neutral_texts, 'Neutral')
print("\nNegative\n")
train_lda_and_plot(negative_texts, 'Negative')
```

```python
top_n_words = 10
for i, topic in lda_model.show_topics(num_topics=10, num_words=top_n_words, formatted=False):
    words, weights = zip(*topic)
    plt.figure(figsize=(10, 5))
    plt.barh(range(len(words)), weights, align='center')
    plt.yticks(range(len(words)), words)
    plt.gca().invert_yaxis()
    plt.xlabel('Weight')
    plt.title(f'Topic {i+1}')
    plt.show()
```

```python
import numpy as np

cv = CountVectorizer(ngram_range=(2,2))
bigrams = cv.fit_transform(data['Preprocessed'])

count_values = bigrams.toarray().sum(axis=0)
ngram_freq = pd.DataFrame(sorted([(count_values[i], k) for k, i in cv.vocabulary_.items()], reverse = True))
ngram_freq.columns = ["frequency", "ngram"]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(ngram_freq.ngram.iloc[:10], ngram_freq.frequency.iloc[:10], color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Bigram Words')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
